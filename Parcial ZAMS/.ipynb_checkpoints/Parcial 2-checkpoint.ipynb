```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
x45 = np.loadtxt("Estrella(45)1.txt")
Capa = []; M45 = []; Lr45 = []; LP45= []; LT45 = []; Lrho45 = []; LL45 = []  
for i in range(len(x45)):
    Capa.append(x45[i][0])
    M45.append(x45[i][1])
    Lr45.append(x45[i][2])
    LP45.append(x45[i][3])
    LT45.append(x45[i][4])
    Lrho45.append(x45[i][5])
    LL45.append(x45[i][6])
    
x75 = np.loadtxt("estrella75.txt")
Capa = []; M75 = []; Lr75 = []; LP75= []; LT75 = []; Lrho75 = []; LL75 = []  
for i in range(len(x75)):
    Capa.append(x75[i][0])
    M75.append(x75[i][1])
    Lr75.append(x75[i][2])
    LP75.append(x75[i][3])
    LT75.append(x75[i][4])
    Lrho75.append(x75[i][5])
    LL75.append(x75[i][6])
    
x52 = np.loadtxt("estrella52.txt")
Capa = []; M52 = []; Lr52 = []; LP52= []; LT52 = []; Lrho52 = []; LL52 = []  
for i in range(len(x52)):
    Capa.append(x52[i][0])
    M52.append(x52[i][1])
    Lr52.append(x52[i][2])
    LP52.append(x52[i][3])
    LT52.append(x52[i][4])
    Lrho52.append(x52[i][5])
    LL52.append(x52[i][6])
    
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)

ax1.plot(10**np.array(Lr45)/7e10,10**np.array(LP45),"b-",label = "$M = 0.45 M_{\odot} $ ")
ax1.set_title('Presion Vs Radio')
ax2.plot(10**np.array(Lr75)/7e10,10**np.array(LP75),"b-",label = "$M = 7.5 M_{\odot} $ ")
ax2.set_ylabel('P')
ax3.plot(10**np.array(Lr52)/7e10,10**np.array(LP52),"b-",label = "$M = 52 M_{\odot} $ ")
ax3.set_xlabel('$R_{\odot}$')

f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)

plt.xscale('log')
plt.yscale('log')
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)

ax1.legend(loc=3)
ax2.legend(loc=3)
ax3.legend(loc=3)

plt.savefig('Presion_graficas.pdf', format='pdf', dpi=1000)


```

```python

# Datos para hacer los plots de polares de las diferentes estrellas
a45 = np.linspace(0,2*np.pi,len(Lr45))
b45 = np.linspace(10**np.min(np.array(Lr45)),10**np.max(np.array(Lr45)),len(Lr45))#np.array(LL52)**10
p45 = 10**np.array(LP45)
A45, B45 = np.meshgrid(a45,p45)
c45 = B45

a75 = np.linspace(0,2*np.pi,len(Lr75))
b75 = np.linspace(10**np.min(np.array(Lr75)),10**np.max(np.array(Lr75)),len(Lr75))#np.array(LL52)**10
p75 = 10**np.array(LP75)
A75, B75 = np.meshgrid(a75,p75)
c75 = B75

a52 = np.linspace(0,2*np.pi,len(Lr52))
b52 = np.linspace(10**np.min(np.array(Lr52)),10**np.max(np.array(Lr52)),len(Lr52))#np.array(LL52)**10
p52 = 10**np.array(LP52)
A52, B52 = np.meshgrid(a52,p52)
c52 = B52

#Modulo que pinta los objetos polares
cmap = plt.cm.Greens
f, axarr = plt.subplots(1,3, figsize=(13,5),subplot_kw=dict(projection='polar'))
st = f.suptitle("Relacion de el radio y la Presion", fontsize="x-large")
st.set_y(0.95)

axarr[0].contourf(a45, b45, c45, cmap = cmap)
axarr[0].set_title('$M = 0.45 M_{\odot} $ ')
axarr[0].set_xticklabels([])

axarr[1].contourf(a75, b75, c75, cmap = cmap)
axarr[1].set_title('$M = 7.5 M_{\odot} $ ')
axarr[1].set_xticklabels([])

axarr[2].contourf(a52, b52, c52, cmap = cmap)
axarr[2].set_title('$M = 52 M_{\odot} $ ')
axarr[2].set_xticklabels([])

plt.setp(axarr[0].get_xticklabels(),visible = False)
plt.setp(axarr[1].get_xticklabels(),visible = False)
plt.setp(axarr[2].get_xticklabels(),visible = False)


#Configuracion de la colorbar

f.subplots_adjust(left=0.05,right=0.85)
cNorm = mpl.colors.Normalize(np.min(np.array(Lr45))**10, np.max(np.array(LP52)**10))
ax3 = f.add_axes([0.91, 0.1, 0.03, 0.8])
cb1 = mpl.colorbar.ColorbarBase(ax3,cmap=cmap,norm = cNorm)
plt.savefig("Polar_presion.pdf", format='pdf', dpi=1000)
```
